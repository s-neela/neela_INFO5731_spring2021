# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

````
covid_tweets['tweets']=covid_tweets['tweets'].str.replace('[!-;:,<>./?@#$%^&*_~{}]','')
covid_tweets['tweets']=covid_tweets['tweets'].str.replace('\d','')
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop=stopwords.words('english')
covid_tweets['tweets']=covid_tweets['tweets'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
covid_tweets['tweets']=covid_tweets['tweets'].apply(lambda x: " ".join(x.lower() for x in x.split()))
from nltk.stem import PorterStemmer
st=PorterStemmer()
covid_tweets['tweets']=covid_tweets['tweets'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
import nltk
from textblob import Word
nltk.download('wordnet')
covid_tweets['tweets']=covid_tweets['tweets'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
covid_tweets.to_csv('Abstract2.csv')
````

For example,  'a phone', 'a laptop' are noun phrases, and 'go-to temple', 'have a party' are verb phrases.
